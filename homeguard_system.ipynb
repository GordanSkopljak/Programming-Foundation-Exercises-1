```python
#=== HomeGuard Security System ===
# Client: The Peterson Family
# ==========================================

import time

# This dictionary tracks the live status of the house
house_state = {
    "system_mode": "HOME",
    "temperature": 70.0,
    "smoke_detected": False,
    "motion_detected": False,
    "door_open": False,
    "door_opened_at": None,
    "is_house_empty": False,
    "ticks_since_last_motion": 0
}

# ==========================================
# Step 2: The Simulation Clock Loop 
# ==========================================
for tick in range(1, 6):
    print(f"\n--- SIMULATION TICK {tick} ---")
    
    # --- Step 2.1: Fake Timeline Events ---
    if tick == 1:
        house_state["motion_detected"] = True
        print("[Event]: Family is moving around getting ready to leave.")
    elif tick == 2:
        house_state["motion_detected"] = False
        print("[Event]: Family closes the door and walks away. House is now quiet.")
    elif tick == 3:
        # Keep motion False so the countdown can finish!
        house_state["motion_detected"] = False
    elif tick == 4:
        # An intruder walks in after it auto-armed!
        house_state["motion_detected"] = True
        print("[Event]: An intruder walks past the window!")


   # ==========================================
    # Step 3: The Rules Engine (Alerts & Logic)
    # ==========================================
    
    # 1. Manage the Auto-Arming Timer Logic
    if house_state["system_mode"] == "HOME":
        if house_state["motion_detected"] == False:
            # House is quiet! Add 1 to our countdown timer
            house_state["ticks_since_last_motion"] += 1
            print(f"[Timer Log]: House has been quiet for {house_state['ticks_since_last_motion']} ticks.")
        else:
            # Someone moved! Reset the timer back to 0
            house_state["ticks_since_last_motion"] = 0

        # If the house stays quiet for 2 ticks, auto-arm to AWAY
        if house_state["ticks_since_last_motion"] >= 2:
            house_state["system_mode"] = "AWAY"
            house_state["is_house_empty"] = True
            print("🔔 NOTIFICATION: No activity detected. System automatically changed to AWAY mode!")

    # 2. Intruder Detection Alert
    if house_state["system_mode"] == "AWAY" and house_state["motion_detected"] == True:
        print("🚨🚨 ALARM: INTRUDER DETECTED! Dispatching emergency services to the Peterson home! 🚨🚨")

    # 3. Frozen Pipe Safety Alert
    if house_state["temperature"] < 35:
        print("🚨 ALERT: CRITICAL TEMPERATURE! Risk of frozen pipes!")


    # --- Step 2.2: Print System Status Tracking ---
    print(f"System Mode: {house_state['system_mode']}")
    print(f"Motion Sensor: {house_state['motion_detected']}")
    
    time.sleep(1)

print("\n--- Simulation Finished! ---")
```
