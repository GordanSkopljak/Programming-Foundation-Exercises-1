```python
#=== HomeGuard Security System ===
# Client: The Peterson Family
# ==========================================
# The HomeGuard system processes risks through 3 segments:
# Security Alerts, Safety Alerts and Comfort Notifications & Patterns


import time

house_state = {
    "system_mode": "HOME",
    "temperature": 70.0,
    "smoke_detected": False,
    "motion_detected": False,
    "door_open": False,
    "is_house_empty": False,
    "ticks_since_last_motion": 0,
    "ticks_door_open": 0
}

# 6 ticks are perfectly engineered to catch every requirement
for tick in range(1, 7):
    print(f"\n--- SIMULATION TICK {tick} ---")
    
    # --- Step 2.1: Precise Target Timeline Events ---
    if tick == 1:
        house_state["motion_detected"] = True
        house_state["temperature"] = 98.0  
        print("[Event]: Family is moving around. The AC breaks down!")
    elif tick == 2:
        house_state["motion_detected"] = False
        house_state["temperature"] = 70.0  
        house_state["door_open"] = True   # Door opens here (Minute 1)
        print("[Event]: Family walks out but leaves the front door wide open!")
    elif tick == 3:
        # Door is open (Minute 2). System auto-arms to AWAY due to quietness.
        print("[Event]: House remains quiet.")
    elif tick == 4:
        # Door is open (Minute 3). Window motion sensor trips.
        house_state["motion_detected"] = True  
        print("[Event]: Sensor detects motion outside the window.")
    elif tick == 5:
        # Door is open (Minute 4). Window motion goes quiet.
        house_state["motion_detected"] = False
        print("[Event]: Window area goes quiet, but door is still hanging open.")
    elif tick == 6:
        # Door is open (Minute 5 -> Triggers Pattern Alert!)
        # Simultaneously, fresh window motion triggers Multi-Sensor Break-In!
        house_state["motion_detected"] = True
        house_state["smoke_detected"] = True 
        house_state["temperature"] = 30.0    
        print("[Event]: CRITICAL! Window motion trips again, smoke breaks out, and temperature plunges!")


    # ==========================================
    # Step 3: The Rules Engine (Alerts & Logic)
    # ==========================================
    
    # Unusual Pattern Tracker (Must evaluate after timeline alterations!)
    if house_state["door_open"]:
        house_state["ticks_door_open"] += 1
    else:
        house_state["ticks_door_open"] = 0

    if house_state["ticks_door_open"] >= 5:
        print(f"⚠️ NOTICE: UNUSUAL PATTERN! Door has been left open for {house_state['ticks_door_open']} minutes!")

    # 1. Auto-Arming Logic
    if house_state["system_mode"] == "HOME":
        if not house_state["motion_detected"]:
            house_state["ticks_since_last_motion"] += 1
            print(f"[Timer Log]: House has been quiet for {house_state['ticks_since_last_motion']} ticks.")
        else:
            house_state["ticks_since_last_motion"] = 0

        if house_state["ticks_since_last_motion"] >= 2:
            house_state["system_mode"] = "AWAY"
            house_state["is_house_empty"] = True
            print("🔔 NOTIFICATION: System automatically changed to AWAY mode!")

    # 2. Security Alerts
    if house_state["system_mode"] == "AWAY":
        if house_state["motion_detected"] and house_state["door_open"]:
            print("🚨🚨 CRITICAL ALARM: MULTIPLE SENSORS TRIGGERED! POSSIBLE BREAK-IN! 🚨🚨")
        elif house_state["motion_detected"]:
            print("🚨 WARNING: Motion detected while away!")
        elif house_state["door_open"]:
            print("🚨 WARNING: Door opened while away!")

    # 3. Smoke Detection Rule
    if house_state["smoke_detected"]:
        print("🔥🧯 CRITICAL ALARM: SMOKE DETECTED! Fire risk confirmed! 🧯🔥")

    # 4. Temperature Safety Rules
    if house_state["temperature"] < 35:
        print("🚨 ALERT: CRITICAL LOW TEMPERATURE! Risk of frozen pipes!")
    elif house_state["temperature"] > 95:
        print("🚨 ALERT: CRITICAL HIGH TEMPERATURE! Risk of equipment failure!")

    # 5. Comfort Range Tracker
    if house_state["system_mode"] == "HOME":
        if house_state["temperature"] < 65 or house_state["temperature"] > 75:
            print(f"⚠️ COMFORT NOTICE: Temperature ({house_state['temperature']}°F) is outside ideal 65-75°F range.")

    print(f"Mode: {house_state['system_mode']} | Motion: {house_state['motion_detected']} | Door: {house_state['door_open']} | Temp: {house_state['temperature']}°F | Smoke: {house_state['smoke_detected']}")
    time.sleep(1)

print("\n--- Simulation Finished! ---")
```
